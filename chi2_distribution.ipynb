```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# 100 linearly spaced numbers
x = np.linspace(0,20,100)

# setting the axes at the centre
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
plt.margins(x=0) #disable plotting margins on x axis
y = chi2.pdf(x,1)
plt.plot(x,y, 'g')
ax.text(1, 0.7, 'n=1', fontsize=12,color='g')
y = chi2.pdf(x,2)
plt.plot(x,y, 'b')
ax.text(2, 0.3, 'n=2', fontsize=12,color='b')
y = chi2.pdf(x,5)
plt.plot(x,y, 'r')
ax.text(5, 0.15, 'n=5', fontsize=12,color='r')
y = chi2.pdf(x,10)
plt.plot(x,y, 'grey')
ax.text(8, 0.15, 'n=10', fontsize=12,color='grey')
#
ax.text(8, 0.7, r'$\chi^2$ prob density', fontsize=14,color='black')
plt.xlabel(r'$\chi^2$')   # latex labels
plt.ylabel(r'$P(\chi^2)$')
# show the plot
plt.show()
fig.savefig("chi2_prob_density.pdf")
```

```python
# cumulative chi2 probs
# 100 linearly spaced numbers
x = np.linspace(0,20,100)

# setting the axes at the centre
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
plt.margins(x=0) #disable plotting margins on x axis
y = chi2.cdf(x,1)
plt.plot(x,y, 'g')
ax.text(1, 0.9, 'n=1', fontsize=12,color='g')
y = chi2.cdf(x,2)
plt.plot(x,y, 'b')
ax.text(1, 0.3, 'n=2', fontsize=12,color='b')
y = chi2.cdf(x,5)
plt.plot(x,y, 'r')
ax.text(2.5, 0.15, 'n=5', fontsize=12,color='r')
y = chi2.cdf(x,10)
plt.plot(x,y, 'grey')
ax.text(6, 0.15, 'n=10', fontsize=12,color='grey')
#
ax.text(8, 0.7, r'cumulative $\chi^2$ prob', fontsize=14,color='black')
plt.xlabel(r'$x$')   # latex labels
plt.ylabel(r'$P(\chi^2<x)$')
# show the plot
plt.show()
fig.savefig("chi2_cumulative_prob.pdf")
```

```python
# cumulative chi2 probs
# 100 linearly spaced numbers
x = np.linspace(0,20,100)

# setting the axes at the centre
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
plt.margins(x=0) #disable plotting margins on x axis
y = chi2.sf(x,1)
plt.plot(x,y, 'g')
ax.text(0.5, 0.1, 'n=1', fontsize=12,color='g')
y = chi2.sf(x,2)
plt.plot(x,y, 'b')
ax.text(2, 0.5, 'n=2', fontsize=12,color='b')
y = chi2.sf(x,5)
plt.plot(x,y, 'r')
ax.text(5.5, 0.4, 'n=5', fontsize=12,color='r')
y = chi2.sf(x,10)
plt.plot(x,y, 'grey')
ax.text(13, 0.3, 'n=10', fontsize=12,color='grey')
#
ax.text(10, 0.85, r'1-cumulative $\chi^2$ prob', fontsize=14,color='black')
plt.xlabel(r'$x$')   # latex labels
plt.ylabel(r'$P(\chi^2>x)$')
# show the plot
plt.show()
fig.savefig("chi2_1-cumulative_prob.pdf")
```
