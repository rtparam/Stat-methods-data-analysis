```python
import random
import matplotlib.pyplot as plt
import numpy as np
mu = 12.0
sigma = 0.5
n_msmts = 5
ntry = 100
list_of_averages = []
fig = plt.figure()
plt.margins(x=0) #disable plotting margins on x axis
for itry in range(ntry):
    msmts = [ random.gauss(mu,sigma) for _ in range(n_msmts)]
    average = sum(msmts)/n_msmts
    list_of_averages.append(average)
    if itry<9: 
        print("itry=",itry,"measurements",msmts,"average",average)
        ax= fig.add_subplot(3,3,itry+1)
        plt.hist(msmts,bins=20,range=(10.,14.))
plt.show()
fig.savefig("several_samples.pdf")
```

```python
fig = plt.figure()
plt.margins(x=0) #disable plotting margins
plt.hist(list_of_averages,bins=20,range=(10.,14.))
plt.show()
fig.savefig("distribution_of_averages.pdf")
print("take standard deviation of distribution as error std-dev=",np.std(list_of_averages))
```
