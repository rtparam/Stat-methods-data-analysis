```python
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
figure, axes = plt.subplots()
Quad = plt.Circle((0,0),1, fill = False)
axes.set_aspect( 1 ) 
axes.add_artist( Quad)
col = []

IT = 10000 

x = np.random.rand(IT)
y = np.random.rand(IT)
f = x**2 + y**2 

Ncirc = 0
Nsq = IT

for i in f:
    if i<1:
        Ncirc+=1 #incrementing circle hit
        col += 'r' # color red if it lies within the circle
    else:
        col += 'b' #color blue if it lies outside the circle
        
plt.scatter(x,y, c=col) 
plt.title('Monte-Carlo calculation of pi')
#estimate for the value of pi based on formula:
pi = 4*Ncirc/Nsq
print("Calculated value of pi = ", pi)
```
